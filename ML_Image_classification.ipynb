```python
#Image classification

#Reading training data

import pandas as pd
df_img=pd.read_csv('d:\\Image_data\\train_image.csv')
df_img.head()
```

```python
#Preparing Training data i.e. X_train

from skimage.io import imread, imshow
from skimage.transform import rescale
import numpy as np

```

```python
#speficy the pysical path of your training images
#specify location of training images on your computer

path='D:\\Image_data\\training\\'

#specify the name of the training image
#Accessing the first image name from train_image.csv file

#here, we are accessing the first record of Image column in df_img dataframe 
file=df_img.iloc[0].Image
path+file
```

```python
#reading the first image from training data
image=imread(path+file, as_gray=True)
imshow(image)

#rescaling the image
img_rescale=rescale(image, 0.25)

#reshaping the image
img_reshape=np.reshape(img_rescale, img_rescale.shape[0]*img_rescale.shape[1])
```

```python
#Applying the above steps to all the training images

data=[]

for i in range(0,33):
    
    #specify location of training images on your computer

    path='D:\\Image_data\\training\\'
    
    #specify the name of the training image
    
    file=df_img.iloc[i].Image
    
    
    #reading the first image from training data
    image=imread(path+file, as_gray=True)
  
    #rescaling the image
    img_rescale=rescale(image, 0.25)

    #reshaping the image
    img_reshape=np.reshape(img_rescale, img_rescale.shape[0]*img_rescale.shape[1])

    data.append(img_reshape)
    print(file)
```

```python
#Preparing X_train

X_train=pd.DataFrame(data)
X_train.to_csv('e:\\training_image.csv')
```

```python
#preparing y_train
y_train=df_img['Name']
y_train.head()
```

```python
#Now preparing testing data
df_test=pd.read_csv('d:\\Image_data\\test_image.csv')
df_test
```

```python
#Applying the above steps to all the testing images

data=[]

for i in range(0,6):
    
    #specify location of testing images on your computer

    path='D:\\Image_data\\test\\'
    
    #specify the name of the training image
    
    file=df_test.iloc[i].Image
    
    
    #reading the first image from training data
    image=imread(path+file, as_gray=True)
  
    #rescaling the image
    img_rescale=rescale(image, 0.25)

    #reshaping the image
    img_reshape=np.reshape(img_rescale, img_rescale.shape[0]*img_rescale.shape[1])

    data.append(img_reshape)
    print(file)
```

```python
#Preparing X_test

X_test=pd.DataFrame(data)
X_test.to_csv('e:\\testing_image.csv')
```

```python
#Preparing y_test

y_test=df_test.Name
y_test
```

```python
#Applying GaussianNB
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train, y_train)
nb.score(X_test, y_test)
```

```python
nb_pred=nb.predict(X_test)
nb_pred
```

```python
#Applying KNeighborsClassifier

from sklearn.neighbors import KNeighborsClassifier 
knn=KNeighborsClassifier()

knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```

```python
knn_pred=knn.predict(X_test)
knn_pred
```
