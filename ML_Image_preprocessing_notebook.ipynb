```python
#Image processing in python

#skimage: pkg for image processing

#imread(): used to read image
#imshow(): used to show images

from skimage.io import imread, imshow

#reading an image using imread()

#syntax: var_name=imread('image_path')

img=imread('d:\\MY.jpg')

#showing the image

imshow(img)
```

```python
#to view the shape of the image

img.shape
```

```python
#here, 1006 is the height of the image or no. of vertical pixels

#768 is the width of the image or no. of horizontal pixels

#3 is the color value of each pixel

print('Resolution of this image:',1006*768*3)
```

```python
print('Height of the image:', img.shape[0], 'pixels')
print('Width of the image:', img.shape[1], 'pixels')
print('No. of colors per pixels:', img.shape[2])

print('Resolution:', img.shape[0]*img.shape[1]*img.shape[2])
```

```python
#To view the color value at a specific pixel

img[600][600]
```

```python
#Here, R=151
#G=67
#B=83
#The values of RGB is between 0-255

img[500][500]
```

```python
#Image classification can not be done on the basis of color

#Reading the image as gray i.e. excluding the color information from the image

df_gray=imread('d:\\MY.jpg', as_gray=True)
imshow(df_gray)
```

```python
df_gray.shape
```

```python
print('Height of the image:', df_gray.shape[0], 'pixels')
print('Width of the image:', df_gray.shape[1], 'pixels')
print('Resolution:', df_gray.shape[0]*df_gray.shape[1])
```

```python
#image rescaling: making image smaller

from skimage.transform import rescale

#rescale(): used to rescale the image i.e. reducing the size of the image
#Syntax: rescale(image_name, rescaling_factor)

#Making the image smaller

df_rescale=rescale(df_gray, 0.25)
imshow(df_rescale)
```

```python
df_rescale.shape
```

```python
print('Height of the image:', df_rescale.shape[0], 'pixels')
print('Width of the image:', df_rescale.shape[1], 'pixels')
print('Resolution:', df_rescale.shape[0]*df_rescale.shape[1])
```

```python
import pandas as pd

df_res=pd.DataFrame(df_rescale)
df_res.to_csv('E:\\image_rescale.csv')
```

```python
#Image reshaping

#reshaping the image to (1, 48384)

import numpy as np

df_reshape=np.reshape(df_rescale, df_rescale.shape[0]*df_rescale.shape[1])
```

```python
df_re=pd.DataFrame(df_reshape)
df_re.to_csv('e:\\image_reshape.csv')
```
