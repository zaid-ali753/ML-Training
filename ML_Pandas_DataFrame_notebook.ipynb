```python
#Pandas DataFrame
import pandas as pd
```

```python
#Empty dataframe
df=pd.DataFrame()
print(df)
print(type(df))
```

```python
#dataframe using list

d=[10,20,30,40,50]
df1=pd.DataFrame(d)
df1
```

```python
#Dataframe with column name

d1=[['Ravi', 50000], ['Sachin', 45000], ['Alok', 39000]]
df2=pd.DataFrame(d1)
df2
```

```python
#Dataframe with column name

d1=[['Ravi', 50000], ['Sachin', 45000], ['Alok', 39000]]
df2=pd.DataFrame(d1, columns=['Name', 'Salary'])
df2
```

```python
#Creating DataFrame using Dictionary

new={'Item':['A', 'B', 'C', 'D', 'E'], 'Price':[600,650,700,850,950]}
df_new=pd.DataFrame(new)
df_new
```

```python
#Data access in DataFrame
print('Price of 3rd Item:', df_new['Price'][2])
```

```python
print('Name of 2nd Item:', df_new['Item'][1])
```

```python
print(df_new['Price'])
```

```python
#Dataframe updation

#to change the Price of 2nd item
df_new['Price'][1]=750
df_new
```

```python
#Adding a new row in a DataFrame
#append():

f2=pd.DataFrame([['F', 1000], ['G', 1050]], columns=['Item', 'Price'])
f2
```

```python
df_new=df_new.append(f2, ignore_index=True)
df_new
```

```python
#Changing name of the columns

#Change the name of the column 'Item' to 'Product_name' and 'Price' to 'Base_price'

df_new.rename(columns={'Item':'Product_name', 'Price':'Base_price'}, inplace=True)
df_new
```

```python
#Column insertion
#We can easily insert a new column at the end of the dataframe as a series

#Adding a new column named 'Unit' in df_new

df_new['Unit']=pd.Series([10, 15, 20, 25, 14, 12, 22])
df_new
```

```python
#Adding a new column using existing columns of the dataframe

#Adding a new column 'Total_price' in df_new

df_new['Total_price']=df_new['Base_price']*df_new['Unit']
df_new
                                                
```

```python
#inserting a column at a specified index in a DataFrame

new_val=['NewDelhi','Mumbai','Chennai', 'Hyderabad','Lucknow', 'Kanpur', 'Bangalore']

df_new.insert(1,column='Place',value=new_val)
df_new
```

```python
#Querying the data from the dataframe

#Accessing the rows where Product_name==A
df_new.loc[df_new['Product_name']=='A']
```

```python
#Accessing data where unit>15
df_new.loc[df_new['Unit']>15]
```

```python
#Accessing data where unit<15
df_new.loc[df_new['Unit']<15]
```

```python
#Accessing data where Total_price>15000
df_new.loc[df_new['Total_price']>15000]
```

```python
#Accessing data where Total_price<15000
a=df_new.loc[df_new['Total_price']<15000]
a
```

```python
#Accessing data where Place==Mumbai
df_new.loc[df_new['Place']=='Mumbai']
```

```python
#Accessing multiple values
df_new.loc[df_new['Place'].isin(['Lucknow', 'Kanpur', 'Mumbai'])]
```

```python
df_new
```

```python
#Accessing data using iloc()
#Used to acess the data using default index
#Syntax: dataframe_name.iloc[row_index, col_index]

#to print the total_price of 3rd row

df_new.iloc[2,4]
```

```python
#To print the place of 4th product

df_new.iloc[3,1]
```

```python
#Syntax: dataframe_name.iloc[start_row_index: end_row_index, start_col_index:end_col_index]

#To access all records of 3rd column
df_new.iloc[:,2]
```

```python
#To access row_index=2 to row_index=5 and all columns
df_new.iloc[2:6, :]
```

```python
#To access row_index=1 to row_index=6 and columns_index=1 to col_index=4
df_new.iloc[1:7, 1:5]
```

```python
#Column deletion
#to delete 'Place' column from df_new

del df_new['Place']
df_new
```

```python
#Row deletion
#Using drop()

#to delete row at index=2

df_new.drop(2, inplace=True)
df_new

```
